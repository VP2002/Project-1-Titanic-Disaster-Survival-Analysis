```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
import pandas as pd
file_handler = open("Employee_Salary_Dataset.csv", "r")
df = pd.read_csv(file_handler, sep=",")
file_handler.close()
gender = {'Male':1,'Female':0}
df.Gender = [gender[item] for item in df.Gender]
print(df)
```

```python
# multiple linear regression

#Setting the parameters
#Setting the value for X and Y

x = df[['Age','Experience_Years']]
y = df['Salary']
```

```python
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.2, random_state=10)
```

```python
mlr= LinearRegression() 
mlr.fit(x_train, y_train)
```

```python
#Printing the model coefficients
print(mlr.intercept_)

# pair the feature names with the coefficients
list(zip(x, mlr.coef_))
```

```python
#Predicting the Test and Train set result 
y_pred_mlr= mlr.predict(x_test) 
x_pred_mlr= mlr.predict(x_train) 
```

```python
y_pred_mlr
```

```python
#Actual value and the predicted value

mlr_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': y_pred_mlr})
mlr_diff
```

```python
print("Prediction for test set: ",y_pred_mlr)
```

```python
# r2 

print('R squared value of the model: {:.2f}'.format(mlr.score(x_test,y_test)*100))
```

# Handling Missing Values

For age

```python
# replacing null values by mean of age

train['Age']= train['Age'].fillna(train['Age'].mean())
test['Age']= test['Age'].fillna(test['Age'].mean())
```

```python
print(train['Age'].isnull().sum())
print(test['Age'].isnull().sum())
```

For Cabin





```python
# printing the values which are not null

train['Cabin'][train['Cabin'].notnull()].head()
```

```python
test['Cabin'][test['Cabin'].notnull()].head()
```

```python
train["Cabin"] = pd.Series([i[0] if not pd.isnull(i) else 'M' for i in train['Cabin']])
test["Cabin"] = pd.Series([i[0] if not pd.isnull(i) else 'M' for i in test['Cabin']])
```

```python
print(train["Cabin"].isnull().sum())
print(test["Cabin"].isnull().sum())
```

```python
test['Cabin'].value_counts()
```

For Embarked

```python
train['Embarked'].value_counts()
```

```python
# fiiling missing values with mode

train['Embarked']= train['Embarked'].fillna(train['Embarked'].mode()[0])
```
